```python
from googleapiclient.discovery import build
```

```python
# API key của bạn
API_KEY = "AIzaSyBptyww2vVmH-BWBVtBTsX9I43cRxbFm5Y"
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"
CHANNEL_ID = "UCrRH2zbxO5LDN82zoY4xWxg"  # ID kênh YouTube
```

```python
# Tạo kết nối API
youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=API_KEY)
```

```python
# Lấy danh sách video từ kênh
def get_video_links(channel_id):
    video_links = []
    next_page_token = None

    while True:
        # Gọi API để lấy danh sách video
        request = youtube.search().list(
            part="id",
            channelId=channel_id,
            maxResults=50,
            pageToken=next_page_token
        )
        response = request.execute()

        # Lưu URL của các video
        for item in response.get("items", []):
            if item["id"]["kind"] == "youtube#video":
                video_id = item["id"]["videoId"]
                video_links.append(f"https://www.youtube.com/watch?v={video_id}")

        # Kiểm tra trang tiếp theo
        next_page_token = response.get("nextPageToken")
        if not next_page_token:
            break

    return video_links
```

```python
# Gọi hàm và in kết quả
video_links = get_video_links(CHANNEL_ID)
for link in video_links:
    print(link)
```
