```python
import youtube_dl
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# Lấy danh sách video từ kênh YouTube dựa trên channelId
def get_video_ids_from_channel(channel_id):
    ydl_opts = {
        'quiet': True,
        'extract_flat': True,  # Chỉ lấy danh sách video mà không tải video
    }
    
    channel_url = f"https://www.youtube.com/channel/UCrRH2zbxO5LDN82zoY4xWxg/videos"  # URL kênh với channelId
    with youtube_dl.YoutubeDL(ydl_opts) as ydl:
        result = ydl.extract_info(channel_url, download=False)
        video_ids = [entry['id'] for entry in result['entries']]
    return video_ids
```

```python
# Tải và in transcript cho video
def print_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        
        # In ra transcript của video
        print(f"\nTranscript cho video {video_id}:\n")
        for entry in transcript:
            start_time = entry['start']
            duration = entry['duration']
            text = entry['text']
            print(f"[{start_time:.2f} - {start_time + duration:.2f}] {text}")
    
    except Exception as e:
        print(f"Không thể tải transcript cho video {video_id}: {e}")
```

```python
# Main function để xử lý
def main():
    channel_id = "UCrRH2zbxO5LDN82zoY4xWxg"  # Channel ID của kênh
    video_ids = get_video_ids_from_channel(channel_id)
    
    for video_id in video_ids:
        print_transcript(video_id)

if __name__ == "__main__":
    main()
```
